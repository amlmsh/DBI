### Eine mögliche Lösung

#### zu 1.)Beschreiben Sie den Lösungsweg analytisch.

$$
    \frac{1}{f} = \frac{1}{|z_b|} + \frac{1}{|z_g|}
$$

$$
    \frac{1}{f} - \frac{1}{|z_g|} = \frac{1}{|z_b|}
$$

$$
    \frac{1}{|z_b|} = \frac{|z_g| - f}{f\cdot|z_g|}
$$

$$
    |z_b| = \frac{f\cdot|z_g|}{|z_g| - f},   ~~~~ |z_g| > f
$$


#### zu 2.) Schreiben Sie eine Routine zur Berechung der Brennweite $f$

##### Eingabe

```python
f = 0.2
zg = 0.5
```

##### Berechnung

```python
zb = (f*zg)/(zg - f)
```
