<b>Schritt 2: Warum ist das Farbbild so bl√§ulich?</b><br>

<b>Antwort:</b> Die Bilddaten werden mittels openCV-Funktionen eingelesen. openCV strukturiert die Farbkan√§le in der Reihenfolge GBR. Die Python-Funktionen setzen die Reihenfolge der Farbkan√§le als RGB voraus. Daher werden die Frabdaten bei der Darstellung falsch interpretiert.

<b>L√∂sung:</b> Es wird ein neues Bild <em>imageRGB</em> erzeug, wo die Farbkan√§le in der RGB-Reihenfolge abgelegt werden.


```python
imageRGB = image.copy()
```

```python
plt.imshow(imageRGB)
```

```python
rows,cols,c = image.shape
```

```python
for i in range(rows):
    for j in range(cols):
        imageRGB[i,j,2] = image[i,j,0] # Python RGB, openCV BGR
        imageRGB[i,j,1] = image[i,j,1]
        imageRGB[i,j,0] = image[i,j,2]
```

```python
plt.imshow(imageRGB)
```

Jetzt werden die Frabdaten korrekt interpretiert.
