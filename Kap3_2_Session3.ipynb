## Session 3: Bilderzeugung

```python
import numpy as np
from glob import glob
from matplotlib import pyplot as plt

import cv2
```

<b>Schritt 1: Definiere die Dimensionen des Bildes</b>

```python
height = 480 # number of rows
width  = 640 # number of columns
```

<b>Schritt 2: Erzeuge ein Grauwertbild mit 8-Bit und den Wertebereich [0...255]</b>

```python
grayImg = np.zeros((height,width),np.uint8)
```

```python
grayImg.shape
```

```python
grayImg
```

```python
plt.imshow(grayImg)
```

```python
plt.imshow(grayImg,cmap='gray')
```

<b>Schritt 3: Erzeuge ein Frabbild</b>

```python
colorImg = np.zeros((height,width,3),np.uint8)
```

```python
colorImg.shape
```

```python
colorImg
```

```python
plt.imshow(colorImg)
```

<b>Schritt 4: Erzeuge zufällige Grauwerte und Farbwert in den beiden Bildern.</b>

```python
import random 

rows,cols = grayImg.shape
for i in range(rows):
    for j in range(cols):
        grayImg[i,j] = random.randint(0,255)
```

```python
plt.imshow(grayImg,cmap='gray')
```

```python
rows,cols,c = colorImg.shape
for i in range(rows):
    for j in range(cols):
        colorImg[i,j,0] = random.randint(0,255)
        colorImg[i,j,1] = random.randint(0,255)
        colorImg[i,j,2] = random.randint(0,255)
```

```python
plt.imshow(colorImg)
```
