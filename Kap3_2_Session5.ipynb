## Session 5: Primitive Objektzeichnung mit openCV 

```python
import numpy as np
from glob import glob
from matplotlib import pyplot as plt

import cv2
```

```python
imgData = glob('../../../pics/imagingbook-images-de3/ch16/*.*')
```

```python
img = cv2.imread(imgData[3]) 
```

```python
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(img, cmap='gray')
```

<b>Schritt 1: Eigene primitiv geometrische Zeichnungen erzeugen.</b>

<b>Schritt 2: Erzeuge ein Bild zum Zeichnen</b>

```python
h, w = img.shape
drwImg = np.zeros((h,w),np.uint8)
```

```python
plt.imshow(drwImg, cmap='gray')
```

<b>Schritt 3: Zeichne einen Kreis.</b>

```python
cv2.circle(drwImg, (np.uint64(w/2),np.uint64(h/3)),50,(255,255,255),-1) 
# arguments: image, center, radius, gray value, thickness
plt.imshow(drwImg, cmap='gray')
```

<b>Schritt 4: Zeichne ein Rechteck.</b>

```python
cv2.rectangle(drwImg,(100,200),(200,300),(255,255,255),-1) 
# arguments: image, center, radius, gray value, thickness
plt.imshow(drwImg, cmap='gray')
```

<b> Schritt 5: √úberlagere beide Bilder um eine ROF (Region of interest) anzuzeigen.</b> 

```python
rows,cols = img.shape
for i in range(rows):
    for j in range(cols):
        if(drwImg[i,j] < 255):
            img[i,j] = 0
```

```python
plt.imshow(img, cmap='gray')
```
