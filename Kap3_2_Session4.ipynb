## Session 4: Skalieren und speichern von Bilddaten

```python
import numpy as np
from glob import glob
from matplotlib import pyplot as plt

import cv2
```

```python
import random  
height = 48 
width  = 64
colorImg = np.zeros((height,width,3),np.uint8)
rows,cols,c = colorImg.shape
for i in range(rows):
    for j in range(cols):
        colorImg[i,j,0] = random.randint(0,255)
        colorImg[i,j,1] = random.randint(0,255)
        colorImg[i,j,2] = random.randint(0,255)
```

```python
colorImg.shape
```

```python
plt.imshow(colorImg)
```

<b>Schritt 1: Skaliere das Bild von (48,64) auf (480, 640)</b>

```python
reszImg1 = cv2.resize(colorImg, (640,480)) #(width,height) oder (columns, rows)!!!
```

```python
reszImg1.shape
```

```python
plt.imshow(reszImg1)
```

<b>Schritt 2: Skaliere das Bild nochmals von (48,64) auf (480, 640), diesemal jedoch mit einem anderen Verfahren.</b>

```python
reszImg2 = cv2.resize(colorImg, (640,480),interpolation = cv2.INTER_NEAREST)
```

```python
plt.imshow(reszImg2)
```

<b>Schritt 3: Skaliere das Bild zu dritten Mal von (48,64) auf (480, 640), diesemal wieder mit einem anderen Verfahren.</b>

```python
reszImg3 = cv2.resize(colorImg, (640,480),interpolation = cv2.INTER_LANCZOS4)
```

```python
plt.imshow(reszImg3)
```

<b> Schritt 4: Schreibe dir urspr√ºngliche Datei in Dateien verschiedener Bilddateiformate.</b>

```python
cv2.imwrite('../../../test.png',reszImg1)
cv2.imwrite('../../../test.tif',reszImg2)
cv2.imwrite('../../../test.jpg',reszImg3)
```
