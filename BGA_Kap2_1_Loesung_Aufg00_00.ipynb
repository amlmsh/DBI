### Eine mögliche Lösung

#### zu 1.)Beschreiben Sie den Lösungsweg analytisch.

##### (1) Finde die Parallele gerade $g_p$, welche durch das Loch verläuft

$g_p$ ist einfach definiert durch den gleichen Richtungsvektor $\vec{u}$ und den Ortsvektor $\vec{p_p}$. Der Wert des Ortsvektors ist trivial, denn der Koordinatenursprung $(0,0,0,)$ ist der Ort des Loches.

Die Gerade $g_p$ hat also wie folgt beschrieben: $\vec{x} = (0,0,0) + s\cdot\vec{u} = s\cdot\vec{u}$.

##### (2) Berechung des Schnittpunkts mit der Bildebene

Die Bildebene ist definiert als alle Punkte $(x,y,z)$ für die gilt $z= -a$. Der Schnittpunkt der Geraden $g_p$ mit der Bildebene ist also jener, wo für die Gerade gilt $z = -a$.

$$  \left( \begin{array}{r} x \\ y \\ -a \end{array}\right)  = 
     \left( \begin{array}{r} 0 \\ 0 \\ 0 \end{array}\right) +
     s \cdot \vec{u} =
     s \cdot \left( \begin{array}{r} x_u \\ y_u \\ z_u \end{array}\right)
$$

$$
     \left( \begin{array}{r} x \\ y \\ -a \end{array}\right)  = 
     \left( \begin{array}{r} s \cdot x_u \\ s \cdot y_u \\ s \cdot z_u \end{array}\right)    
$$

Daraus ergeben sich drei Gleichungen und leicht ist zu sehene, dass $s = \frac{-a}{z_u}$. Damit erhalten wir  nun ganz einfach die folgenden Gleichungen für die Berechnung von $x$ bzw. $y$:

$$
    x =  - \frac{a\cdot x_u }{z_u}
$$

$$
    y =  -\frac{a\cdot y_u }{z_u}
$$

#### zu 2.) Schreiben Sie eine Routine zur Berechung des Fluchtpunkts.

##### Eingabe

```python
x_u = -1
y_u = 0
z_u = 1
a   = 7
```

##### Berechung

```python
x = (-a*x_u) / z_u
y = (-a*y_u) / z_u
```
